For KNN we need to convert categorical attributes to numerical values, as performed next

```python
def numericalize(df, col_name, new_col_name):
    df[new_col_name] = df[col_name].astype('category')
    df[[new_col_name]] = df[[new_col_name]].apply(lambda x: x.cat.codes)

numericalize(df_final_train_abt, 'Gender', 'GenderNum')
numericalize(df_final_train_abt, 'County', 'CountyNum')
numericalize(df_final_train_abt, 'IncomeGroup', 'IncomeGroupNum')
numericalize(df_final_train_abt, 'HeldLoanPreviously', 'HeldLoanPreviouslyFlag')

df_final_train_abt.dtypes
df_final_train_abt
```

## Create training and testing data
To have equal raito of training and testing data with/without LoanFlag we have to split ABT based on target and then assign randomize and finally merge

### Setting

```python
Target = 'LoanFlag'
Features = ['Age', 'GenderNum', 'CountyNum', 'IncomeGroupNum', 'HeldLoanPreviouslyFlag', 'NumberOfProductsInbank', 'AverageTXNAmount', 'NumTransactions', 'gamber', 'luxurious', 'golfer', 'traveler', 'gamer', 'shopper', 'cinephilia', 'car_renter']

np.random.seed(0)
```

## Preprocess Data

```python
random_state = np.random.RandomState(0)


X = df_final_train_abt[Features].values
y = pd.factorize(df_train_abt['LoanFlag'])[0]

kfold = 16
```

## Helper

```python
def train_plot_by_classifier(classifier, kfold, kX, y):
    cv = StratifiedKFold(n_splits=kfold)

    tprs = []
    aucs = []
    mean_fpr = np.linspace(0, 1, 100)

    i = 0
    for train, test in cv.split(X, y):
        probas_ = classifier.fit(X[train], y[train]).predict_proba(X[test])
        fpr, tpr, thresholds = roc_curve(y[test], probas_[:, 1])
        tprs.append(interp(mean_fpr, fpr, tpr))
        tprs[-1][0] = 0.0
        roc_auc = auc(fpr, tpr)
        aucs.append(roc_auc)
        plt.plot(fpr, tpr, lw=1, alpha=0.3, label='ROC fold %d (AUC = %0.2f)' % (i, roc_auc));
        i += 1
        
    plt.plot([0, 1], [0, 1], linestyle='--', lw=2, color='r', label='Luck', alpha=.8);

    mean_tpr = np.mean(tprs, axis=0)
    mean_tpr[-1] = 1.0
    mean_auc = auc(mean_fpr, mean_tpr)
    std_auc = np.std(aucs)
    plt.plot(mean_fpr, mean_tpr, color='b', label=r'Mean ROC (AUC = %0.2f $\pm$ %0.2f)' % (mean_auc, std_auc), lw=2, alpha=.8);

    std_tpr = np.std(tprs, axis=0)
    tprs_upper = np.minimum(mean_tpr + std_tpr, 1)
    tprs_lower = np.maximum(mean_tpr - std_tpr, 0)
    plt.fill_between(mean_fpr, tprs_lower, tprs_upper, color='grey', alpha=.2, label=r'$\pm$ 1 std. dev.');

   
    plt.xlim([-0.05, 1.05])
    plt.ylim([-0.05, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver operating characteristic example')
    plt.legend(loc="lower right", bbox_to_anchor=(1.65, -0.25))
    plt.show()
    
```

# Decision Tree

```python
kfold = 16
decision_tree_clf = DecisionTreeClassifier();
train_plot_by_classifier(decision_tree_clf, kfold, X, y)
```

# k-nearest neighbors

```python
kfold = 16
knn_clf = KNeighborsClassifier();
train_plot_by_classifier(knn_clf, kfold, X, y)
```

## Random Forest

```python
kfold = 16
random_forest_clf = RandomForestClassifier();
train_plot_by_classifier(random_forest_clf, kfold, X, y)
```

## Ada Boost

```python
kfold = 16
ada_boost_clf = AdaBoostClassifier();
train_plot_by_classifier(ada_boost_clf, kfold, X, y)
```

## Ensemble (Voting)

```python
eclf = VotingClassifier(estimators=[('Decision Tree', random_forest_clf),
                                    ('k-nearest neighbors', knn_clf),
                                    ('Random Forest', random_forest_clf),
                                    ('Ada Boost', ada_boost_clf)], 
                                    voting='soft')

train_plot_by_classifier(eclf, kfold, X, y)
```

```python
df_test_abt = pd.read_csv("../specs/clean/Test Sample/TEST - AbastractBaseTable.csv", encoding='latin-1', index_col=False)
df_test_abt.head(3)

df_test_abt.dtypes
```

```python
df_final_test_abt = pd.merge(df_test_abt, df_test_txn, on='ClientID')
df_final_test_abt.head(3)
df_final_test_abt.dtypes
```

```python
numericalize(df_final_test_abt, 'Gender', 'GenderNum')
numericalize(df_final_test_abt, 'County', 'CountyNum')
numericalize(df_final_test_abt, 'IncomeGroup', 'IncomeGroupNum')
numericalize(df_final_test_abt, 'HeldLoanPreviously', 'HeldLoanPreviouslyFlag')

df_final_test_abt.dtypes

X = df_final_test_abt[Features].values
```

```python
result_df = pd.DataFrame(eclf.predict_proba(X))
result_df.columns = ['0', 'Score']
result_df['ClientID'] = df_test_abt['ClientID']

threathold_groups = {
    'VeryLow' : (0, 0.05),
    'Low' : (0.05, 0.15),
    'Medium' : (0.15, 0.30),
    'High' : (0.30, 0.55),
    'VeryHigh' : (0.50, 1.0),
}
def build_group(x):
    for key, val in threathold_groups.items():
        if val[0] <= x <= val[1]:
            return key

result_df['Group'] = result_df['Score'].apply(build_group)

result_df = result_df[['ClientID', 'Score', 'Group']]

pd.DataFrame.to_csv(result_df, "../specs/result/RESULT - Model Result.csv", encoding='utf-8', index=False)
```
